# Pandemic/bacteria population

## Model 1: linear

Let $N_0$ be the number of bacteria in a sample at $t=0$. If the population growth linearly, we find the population $N_{k+1}$ at the generation $k+1$ equals the the population $N_k$ at the generation $k$ plus a constant $r$:
$$
N_{k+1} = N_{k}+r\ .
$$
This means that at the generation $k\geq 0$, the total number of bacteria is:
$$
N_{k}=N_0 + k r\ .
$$

First, let's use a for loop to calculate the number of bacteria after, say, 10 generations. 

```python
u=100 # Initial condition
N=10
r=2
for n in range(N+1):
  print(n,u)
  u=u+r
  n=n+1
```

We could also use the while loop:

```python
u=100
N=10
r=2
n=0
while n<=10:
  print(n,u)
  u=u+r
  n=n+1
```

Or create a list:

```python
u=100
N=10
r=2
L=[]
for n in range(N+1):
  L.append(u)
  u=u+r
  n=n+1
print(L)
```

And graph it:

```python
from matplotlib import pyplot as plt
u=100
N=10
r=2
L=[]
for n in range(N+1):
  L.append(u)
  u=u+r
  n=n+1
plt.plot(L,'ro')


```

##Model 2: Exponential

Now, we consider a more realistic model: the exponential growth. 
Let $N_0$ be the number of bacteria in a sample at $t=0$. The rate of change of the population is constant:
$$
\frac{N_{k+1}-N_k}{N_k} = r\ ,
$$
which can be rewritten as:
$$
N_{k+1} = (1+r)N_k\ ,
$$
This means that at the generation $k\geq 0$, the total number of bacteria is:
$$
N_{k}=N_0 (1+r)^k\ .
$$

We can now adapt the codes we made above. Let's take $N_0=100$ and $r=0.03$ (i.e., 3\% increase per generation) .

```python
u=100
N=10
r=0.03
for n in range(N+1):
  print(n,u)
  u=u*(1+r)
  n=n+1
```

We could also use the while loop:

```python
u=100
N=10
r=0.03
n=0
while n<=10:
  print(n,u)
  u=u*(1+r)
  n=n+1
```

Or create a list:

```python
u=100
N=10
r=0.03
L=[]
for n in range(N+1):
  L.append(u)
  u=u*(1+r)
  n=n+1
print(L)
```

And graph it (we now take 100 generations):

```python
from matplotlib import pyplot as plt
u=100
N=100
r=0.03
L=[]
for n in range(N+1):
  L.append(u)
  u=u*(1+r)
  n=n+1
plt.plot(L,'ro')


```

We can graph both model over 100 generations on the same graph with $r_1=10$ (linear growth) and $r_2=0.03$ (exponential growth) and $N_0=100$ (for both models). 

Notice that these two models are analogous to simple and compound interests. 

```python
from matplotlib import pyplot as plt
u1=100
u2=100
N=100
r1=10
r2=0.03
L1=[]
L2=[]
for n in range(N+1):
  L1.append(u1)
  L2.append(u2)
  u1=u1+r1
  u2=u2*(1+r2)
  n=n+1
plt.plot(L1,'ro'), plt.plot(L2,'bo'), plt.legend(["linear","exponential"])
```

## Model 3: Logistic

The most realistic model is the so-called logistic growth model. It is used to model pandemic growth (e.g., beginning of the covid pandemic) and evolution of chemical component in a chemical reaction (c.f., Phys-Chem_ODEs). The specificity of this model is that the population saturates at a fixed value instead of growing indefinitely.   

Let $N_0$ be the number of bacteria in a sample at $t=0$. The rate of change of the population does now depend on the population itself:
$$
\frac{N_{k+1}-N_k}{N_k} = r\cdot\left(1-\frac{N_{k}}{L}\right)\ ,
$$
where $L$ is a positive number that is supposed to be large compared with $N_0$ (if $L$ is small, the logistic sequence can show some funny pattern somehow related to fractal and chaos, but that's beyond the scope of this training class).
The equation above can be rewritten as:
$$
N_{k+1} = N_k\cdot\left(1+r\cdot\left(1-\frac{N_{k}}{L}\right)\right)\ .
$$
For this model, we don't have a closed-form expression for $N_k$. Therfore, we have to write a code to solve it numerically. 

We can now adapt the codes we made above. Let's take $N_0=100$ and $r=0.03$ (i.e., 3\% increase per generation) and $L=1000$.

```python
u=100
N=10
r=0.03
L=1000
for n in range(N+1):
  print(n,u)
  u=u*(1+r*(1-u/L))
  n=n+1
```

We could also use the while loop:

```python
u=100
N=10
r=0.03
L=1000
n=0
while n<=10:
  print(n,u)
  u=u*(1+r*(1-u/L))
  n=n+1
```

Or create a list:

```python
u=100
N=10
r=0.03
L=1000
Liste=[] # Careful! One has to change L to something else because we use L as parameter. I chose Liste instread of L. 
for n in range(N+1):
  Liste.append(u)
  u=u*(1+r*(1-u/L))
  n=n+1
print(Liste)
```

And graph it (we now take 500 generations):

```python
from matplotlib import pyplot as plt
u=100
N=500
r=0.03
L=1000
Liste=[]
for n in range(N+1):
  Liste.append(u)
  u=u*(1+r*(1-u/L))
  n=n+1
plt.plot(Liste,'ro')


```

Notice that the population saturates at $1000$, which corresponds to the value of $L$. For $L$ large compared to $N_0$, we can show that this is indeed approximatively the case. 

We can graph the three models over 500 generations on the same graph with $r_1=10$ (linear growth) and $r_2=0.009$, $r_3=0.03$ (exponential and logistic growth), $L=1000$ (logistic), and $N_0=500$. 

```python
from matplotlib import pyplot as plt
u1=100
u2=100
u3=100
N=500
r1=10
r2=0.009
r3=0.03
L=1000
L1=[]
L2=[]
L3=[]
for n in range(N+1):
  L1.append(u1)
  L2.append(u2)
  L3.append(u3)
  u1=u1+r1
  u2=u2*(1+r2)
  u3=u3*(1+r3*(1-u3/L))
  n=n+1
plt.plot(L1,'ro'), plt.plot(L2,'bo'), plt.plot(L3,'go'), plt.legend(["linear","exponential","logistic"])
```

# Allelic population

##Model  

Consider two allels: A and a. $N_A$ and $N_a$ denote the population of A and a, respectiely. $N=N_A+N_a$ denotes the total population.  

Let's assume that the allelic populations is modeled by a geometric sequence:  
\begin{equation}
\begin{cases}
N_A(k+1) = q_A \times N_A(k) \\
N_a(k+1) = q_a \times N_a(k)
\end{cases}
\end{equation}
$q_A$ and $q_a$ are the rate of change of the population, and $k=0,1,2,3,\cdots$ is the number of generations. $N_A(0)$ and $N_a(0)$ are the initial populations of A and a, respectiely, and $N(0)=N_A(0)+N_a(0)$ is the initial total population.  

The allelic frequencies are defined as follows:
\begin{equation}
\begin{cases}
f_A(k) = \frac{N_A(k)}{N(k)} \\
f_a(k) = \frac{N_a(k)}{N(k)}
\end{cases}\ ,
\end{equation}
where the total number of allels is:
\begin{equation}
N(k) = N_A(k)+N_a(k)\ .
\end{equation}

If $q_A>qa$, we say that the allel A has a selective advantage. We expect that the frequency of A increases and approaches 1 as k increases, while the frequency of a decreases and approaches 0. 

In what follows, we will check whether this assumption is correct. 

## Code

```python
from matplotlib import pyplot as plt
def list(qA,qa,n,NA,Na):
    uA=NA
    ua=Na
    u=uA+ua
    LA=[]
    La=[]
    for k in range(0,200):
        LA.append(uA/u)
        uA=qA*uA
        La.append(ua/u)
        ua=qa*ua
        u=uA+ua
        k=k+1
    return plt.plot(LA,'rx'),plt.plot(La,'bx'),plt.legend(["Allele A", "Allele a"])
list(1.05,1.01,200,400,600)
```

## Random model

### Model

In this model, we consider that the rates $q_A$ and $q_a$ are fluctuating randomly:
$$
q_{A}(k) = q_{A} + \lambda_A\cdot X_A, q_{A}(a) = q_{a} + \lambda_a\cdot  X_a\ ,
$$
where $X_A$ and $X_a$ are two random variables (homogeneous between two values min and max) and $\lambda_A, \lambda_a$ are the amplitude of the fluctuations.

To modify the code, we have to import the function *random* and generate random numbers using the command *random.uniform(min,max)*. 

```python
from matplotlib import pyplot as plt
import random
def list(qA,qa,lA,la,min,max,n,NA,Na):
    uA=NA
    ua=Na
    u=uA+ua
    LA=[]
    La=[]
    for k in range(0,200):
        LA.append(uA/u)
        uA=(qA+lA*random.uniform(min,max))*uA
        La.append(ua/u)
        ua=(qa+la*random.uniform(min,max))*ua
        u=uA+ua
        k=k+1
    return plt.plot(LA,'rx'),plt.plot(La,'bx'),plt.legend(["Allele A", "Allele a"])
list(1.05,1.01,0.4,0.4,0,1,200,400,600)
```

## Dicrete model

```python
import matplotlib.pyplot as plt
import math as m

def predator_prey_eq(y,z,n,a,b,c,d):
    y_liste=[]
    z_liste=[]
    for k in range(1,n+1):
        z_liste.append(z)
        y_liste.append(y)
        y=y+(a*y-b*y*z)
        z=z+(c*y*z-d*z)
        k=k+1
    return plt.plot(y_liste), plt.plot(z_liste), plt.legend(["Prey", "Predator"])

predator_prey_eq(10,10,1000,0.11,0.04,0.01,0.04)
```

## Continuous model

Consider two species A and B, A is the prey and B the predator. The function $y$ model the evolution of A as a function of time $t$ and $z$ model the evolution of B. The equations satisfied by the two functions $y$ and $z$ are two ordinary differential equations:
$$
(S):\begin{cases}
\frac{dy}{dt} = a\ y-b\ y\ z \\
\frac{dz}{dt} = c\ y\ z-d\ z
\end{cases}\ ,
$$
where $a,b,c,d$ are four positive real numbers. We can easily interprete this equation. If there were no predator, the population of prey will grow exponentially fast $\frac{dy}{dt} = a\ y,\ a>0$ (we assume they have an unlimited amount of food, like vegetable). On the other hand, if there were no prey, the population of predator will decrease exponentially fast $\frac{dz}{dt} =-d\ z,\ d>0$  (no food, so they die). Now, if we introduce a population of prey, then the rate of change of predator grows as the population of prey grows: $\frac{dz}{dt} = (c\ y-d)z,\ c,d>0$. Similarly, the population of prey decreases as the the population of predator grows: $\frac{dy}{dt} = (a-b\ z)\ y,\ a,b>0$. 

In order to solve this system of equations (S), we can discretize the time interval $[0,t_f]$ into small equal size interval of length $h = \frac{t_f}{n}$, where $n$ is large enough so that $h\ll t_f$. We denote $y_k = y(t_k)$ and $z_k = z(t_k)$, where $t_k = k\cdot h$, then:
$$
(S_n):\begin{cases}
\frac{y_{k+1}-y_{k}}{h} = a\ y_k-b\ y_k\ z_k \\
\frac{z_{k+1}-z_{k}}{h}= c\ y_k\ z_k-d\ z_k
\end{cases}\ ,
$$
and where we fix the initial conditions: $y(0)=y_0,\ z(0)=z_0$. 

We can rewrite $(S_n)$ as a system of recursive formulae:
$$
(S_n):\begin{cases}
y_{k+1} = y_k + h\cdot(a\ y_k-b\ y_k\ z_k) \\
z_{k+1} = z_k + h\cdot(c\ y_k\ z_k-d\ z_k)
\end{cases}\ ,
$$
and use similar algorithm as above to solve it. 

Below, we write a code to solve these equations for the following set of parameters: $t_i=0,\ y_0=10,\ z_0=10,\ h=0.1,\ n=1000,\ a=1.1,\ b=0.4,\ c=0.1,\ d=0.4$. It turns out that we fixed $h$ and $n$, so the final time is $t_f = h\cdot n = 100$.  

```python
import matplotlib.pyplot as plt
import math as m

def predator_prey_eq(t,y,z,h,n,a,b,c,d):
    t_liste=[]
    y_liste=[]
    z_liste=[]
    for k in range(1,n+1):
        z_liste.append(z)
        y_liste.append(y)
        t_liste.append(t)
        y=y+h*(a*y-b*y*z)
        z=z+h*(c*y*z-d*z)
        t=t+h
        k=k+1
    return plt.plot(t_liste, y_liste), plt.plot(t_liste, z_liste), plt.legend(["Prey", "Predator"])

predator_prey_eq(0,10,10,0.1,1000,1.1,0.4,0.1,0.4)
```
