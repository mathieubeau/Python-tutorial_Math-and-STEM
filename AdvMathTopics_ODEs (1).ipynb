# Numerical solution of an ODE: Euler method

Consider an ODE:
$$
y'=-y\\
y(0)=1
$$


```python
from matplotlib import pyplot as plt

L=[]
y=1
xi=0
xf=1
n=100
h=(xf-xi)/n
for k in range(n+1):
  L.append(y)
  y=(1+h)*y

print(L)


```

# System of ODEs

Consider the system of coupled ODEs:
$$
\begin{cases}
  \frac{dx}{dt}=0.2x-y\\
  \frac{dy}{dt}=x+0.1y
\end{cases}
$$
with
$$
\begin{cases}
  x(0)=1\\
  y(0)=2
\end{cases}
$$


```python
from matplotlib import pyplot as plt
import numpy as np

Lt=[]
Lx=[]
Ly=[]
ti=0
tf=100
xi=1
yi=2
t=ti
x=xi
y=yi
n=1000
h=(tf-ti)/n
for k in range(n+1):
  Lt.append(t)
  Lx.append(x)
  Ly.append(y)
  x=x+h*(0.2*x-y)
  y=y+h*(x-0.1*y)
  t=t+h

plt.plot(Lt,Lx,'r-')
plt.plot(Lt,Ly,'b-')
plt.show()

plt.plot(Lx,Ly,'b-')
plt.show()
```

## Predator-prey equation


Consider the system of coupled ODEs:
$$
\begin{cases}
  \frac{dx}{dt}=\frac{2}{3}x-\frac{4}{3}x y\\
  \frac{dy}{dt}=xy-y
\end{cases}
$$
with
$$
\begin{cases}
  x(0)=0.9\\
  y(0)=1.8
\end{cases}
$$


```python
from matplotlib import pyplot as plt
import numpy as np

Lt=[]
Lx=[]
Ly=[]
ti=0
tf=50
xi=0.9
yi=1.8
t=ti
x=xi
y=yi
n=1000
h=(tf-ti)/n
for k in range(n+1):
  Lt.append(t)
  Lx.append(x)
  Ly.append(y)
  x=x+h*((2/3)*x-(4/3)*y*x)
  y=y+h*(1*x*y-1*y)
  t=t+h

plt.plot(Lt,Lx,'r-')
plt.plot(Lt,Ly,'b-')
plt.show()

plt.plot(Lx,Ly,'b-')
plt.show()
```

## Lorenz system

The equations are:
$$
\begin{cases}
  \frac{dx}{dt} = \sigma(y-x)\\
  \frac{dy}{dt} = x(\rho-z)-y\\
  \frac{dz}{dt} = xy-\beta z\\
\end{cases}\ ,
$$
where $x$ is proportional to the rate of convection, $y$ to the horizontal temperature variation, and $z$ to the vertical temperature variation (see [wiki](https://en.wikipedia.org/wiki/Lorenz_system)).

Write a python code to graph $z$ versus $x$ for $\rho = 28, \sigma = 10, and \beta = \frac{8}{3}$. Use the
following initial conditions: $x(0)=1,\ y(0)=z(0)=0$ and solve the equations for $t\in[0,50]$.

```python
from matplotlib import pyplot as plt
import numpy as np

Lt=[]
Lx=[]
Ly=[]
Lz=[]
sigma = 10
rho = 28
beta = 8/3
ti=0
tf=50
xi=0.3
yi=0
zi=0
t=ti
x=xi
y=yi
z=zi
n=10000
h=(tf-ti)/n
for k in range(n+1):
  Lt.append(t)
  Lx.append(x)
  Ly.append(y)
  Lz.append(z)
  x=x+h*sigma*(y-x)
  y=y+h*(x*(rho-z)-y)
  z=z+h*(x*y-beta*z)
  t=t+h

#plt.plot(Lt,Lx,'r-')
#plt.plot(Lt,Ly,'b-')
#plt.show()

plt.plot(Lx,Lz,'b-')
plt.show()

plt.plot(Lx,Ly,'b-')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

Lt=[]
Lx=[]
Ly=[]
Lz=[]
sigma = 10
rho = 28
beta = 8/3
ti=0
tf=50
xi=0.3
yi=0
zi=0
t=ti
x=xi
y=yi
z=zi
n=10000
h=(tf-ti)/n
for k in range(n+1):
  Lt.append(t)
  Lx.append(x)
  Ly.append(y)
  Lz.append(z)
  x=x+h*sigma*(y-x)
  y=y+h*(x*(rho-z)-y)
  z=z+h*(x*y-beta*z)
  t=t+h

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(Lx, Ly, Lz)
ax.set_xlabel('Lx')
ax.set_ylabel('Ly')
ax.set_zlabel('Lz')
ax.set_title("Equation de Lorenz en 3D")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def lorenz(x, y, z, sigma, rho, beta):
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return dx, dy, dz

def simulate_lorenz(x0, y0, z0, sigma, rho, beta, dt, steps):
    xs = np.zeros(steps)
    ys = np.zeros(steps)
    zs = np.zeros(steps)

    xs[0], ys[0], zs[0] = x0, y0, z0

    for i in range(1, steps):
        dx, dy, dz = lorenz(xs[i-1], ys[i-1], zs[i-1], sigma, rho, beta)
        xs[i] = xs[i-1] + (dx * dt)
        ys[i] = ys[i-1] + (dy * dt)
        zs[i] = zs[i-1] + (dz * dt)

    return xs, ys, zs

# Set parameters
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0
dt = 0.01
steps = 10000

# Set initial conditions
x0, y0, z0 = 0.1, 0.0, 0.0

# Simulate the Lorenz system
xs, ys, zs = simulate_lorenz(x0, y0, z0, sigma, rho, beta, dt, steps)

# Plot the Lorenz attractor in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor')

# Show the plot
plt.show()






```

# Second order ODEs

## Linear equation

Consider the second order linear ODE:
$$
y''=-\omega^2 y -2\beta y'\ ,
$$
with $y(0)=y_0$ and $y'(0)=v_0$ and where $y' = \frac{dy}{dt}$ and $y'' = \frac{d^2y}{dt^2}$.
To solve this equation, we put
$$
z=y'
$$
Hence, the equation above can be written as a system of two coupled linear ODEs:
$$
\begin{cases}
  y' = z \\
  z' = -\omega^2 y - 2\beta z
\end{cases}
$$
with $y(0)=y_0$ and $z(0)=v_0$.


In the code below, we set the values of $\omega$, $\beta$, $x_0$ and $v_0$:

```python
from matplotlib import pyplot as plt
import numpy as np

Lt=[]
Ly=[]
Lz=[]
om=0.2
beta=0.01
ti=0
tf=500
yi=1
zi=0
t=ti
y=yi
z=zi
n=1000
h=(tf-ti)/n
for k in range(n+1):
  Lt.append(t)
  Ly.append(y)
  Lz.append(z)
  z=z+h*(-om**2*y-2*beta*z)
  y=y+h*z
  t=t+h

plt.plot(Lt,Ly,'r-')
plt.plot(Lt,Lz,'b-')
plt.show()

plt.plot(Ly,Lz,'g-')
plt.show()
```
