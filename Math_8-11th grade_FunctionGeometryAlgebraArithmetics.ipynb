## Pythagorean theorem

Consider a triangle ABC. You want to check whether the triangle is a right triangle (in A). For that you check whether $AB^2+AC^2=BC^2$ using a *if else* loop:

```python
a = float(input("AB= "))
b = float(input("AC= "))
c = float(input("BC= "))

if a**2+b**2 == c**2:
    print("ABC is a right triangle")
else:
    print("ABC is a not a right triangle")
```

You can do it using a system of coordinates where $A(x_A,y_A)$, $B(x_B,y_B)$, $C(x_C,y_C)$:

```python
xA = float(input("entree l'abscisse de A"))
yA = float(input("entree l'ordonnée de A"))

xB = float(input("entree l'abscisse de B"))
yA = float(input("entree l'ordonnée de B"))

xC = float(input("entree l'abscisse de C"))
yC = float(input("entree l'ordonnée de C"))

if (xB-xA)**2+(yB-yA)**2+(xC-xA)**2+(yC-yA)**2-(xC-xB)**2-(yC-yB)**2 ==0:
    print("rectangle en A")
else:
    print("non rectangle en A")
```

Now, if you know that ABC in a right trangle and BC is the hypotenuse. You can calculate the length of the hypotenuse if you know the length of the other sides:

```python
import numpy as np # We use the numpy library to calculate the square root

a = float(input("AB= "))
b = float(input("AC= "))
c = float(input("BC= "))

c==np.sqrt(a**2+b**2)
```

## Trigonometry

Consider a right triangle ABC, BC is the hypothenuse. We know the lengths AB, AC, and BC. We can calculate the angle $\hat{B} = \widehat{ABC}$ if we first calculate either the cosine or the sine or the tangent of this angle using the well-known formulae:
$$
\cos(\hat{B}) = \frac{AB}{BC},\ \sin(\hat{B}) = \frac{AC}{BC},\ \tan(\hat{B}) = \frac{AC}{AB}\ .
$$
Once we know their respective value, we can find the angle using the inverse functions: $\arccos$ or $\arcsin$ or $\arctan$ (other notation: $\cos^{-1}, \sin^{-1}, \tan^{-1}$). 

I suggest you use $AB=3,\ AC=4,\ BC=5$ for the example below:

```python
import numpy as np # We use the numpy library to calculate arccos/arcsin/arctan

a = float(input("AB= "))
b = float(input("AC= "))
c = float(input("BC= "))

cosA = a/c
sinA = b/c
tanA = b/a 

print(np.arccos(cosA)*180/np.pi)
print(np.arcsin(sinA)*180/np.pi)
print(np.arctan(tanA)*180/np.pi)


```

```python
import numpy as np

n=20
L=[]
for a in range(1,n+1):
  for b in range(a,n+1):
    c=np.sqrt(a**2+b**2)
    if float(c).is_integer()==True:
      L.append([a,b,c])
      b=b+1
    else:
      b=b+1
  a=a+1
print(L)
```

## Solving linear equations

To solve equations, we need to use sympy to create symbols that play the role of variables (instead of using parameters).

For instance, let's solve the equation:
$$
2x-4=0
$$

```python
from sympy import symbols, solve # we import symbols and solve from sympy

x = symbols('x')
expr = 2*x-4 # This reads y=2x-4
sol = solve(expr) # This is the equation to solve: y=0 <=> 2x-4=0
sol # we print the solution, which is nothing but a vector (or a list in the python terminology)
```

Other method:

```python
from sympy import symbols, Eq, solve

x = symbols("x")
equation = Eq((2*x - 4), 0)
print("Equation:", equation)
solution = solve(equation,x)
print("Solution:")
print(solution)
```

If we want to use the solution as a number (instead of a list) and calculate stuffs with it, we have to extract the number from the list:

```python
a = sol[0]
print(a)
b=2*a+5
print(b)
```

## Solving systems of linear equations

We can solve a system of linear equations, for example:
$$
\begin{cases}
2x+4y=10\\
4x+2y=30
\end{cases}
$$

```python
from sympy import symbols, Eq, solve

x, y = symbols("x y") # we introduce the two symbols x, y
equation_1 = Eq((2*x + 4*y), 10) # We define the two equations
equation_2 = Eq((4*x + 2*y), 30)
print("Equation 1:", equation_1) # We can print these equations
print("Equation 2:", equation_2)
solution = solve((equation_1, equation_2), (x, y)) # we solve the system of equations
print("Solution:", solution) # we print the solutions giving the values of x and y
print(solution)
```

# Function

## Defining a function

Functions are very important in python as they are in mathematics. We can define a function $y=f(x)$ and calculate its value for a given number. For example, take $f(x)=x^2$ and calculate the value $f(2)$:

```python
def f(x):
    y=x**2
    return y
f(2)
```

Or take $g(x)=-x^2+3x-2$ and calculate $g(-3)$

```python
def g(x):
    y=-x**2+3*x-2
    return y
g(-3)
```

Or take $h(x) = \frac{2x-1}{x+1}$ and calculate $h(7)$

```python
def h(x):
    y=(2*x-1)/(x+1)
    return y
h(7)
```

Or take $i(x) = e^{x}$ and $j(x)=\ln(x)$ and calculate the following quantities: 


*  $i(1)$
*  $j(2)$
* sum: $i(1)+j(2)$
* product: $i(1)\times j(2)$
* quotient: $\frac{i(1)}{j(2)}$
* power: $i(1)^{j(2)}$ and $j(1)^{i(2)}$
* composition: $i(j(2))$ and $j(i(1))$

```python
import numpy as np # we import  numpy to use the exponential function and the logarithmic function
def i(x):
    y=np.exp(x)
    return y

def j(x):
    y=np.log(x)
    return y

print(i(1))
print(j(2))
print(i(1)+j(2))
print(i(1)*j(2))
print(i(1)/j(2))
print(i(1)**j(2),j(2)**i(1))
print(i(j(2)),j(i(1)))
```

We can use these functions in a code whenever we want to use them once they have been defined. For instance, imagine we want to find the least integer $m$ for which $e^{-m}$ is less than $0.0001$:

```python
def k(x):
    y=np.exp(-x)
    return y
m=0
while k(m)>=0.0001: # we calculate the value of k(m) and continue interate if k(m)>=0.01
    m=m+1
print("m= ",m," exp(m)= ",k(m)," exp(m-1)= ",k(m-1)) # we print the value of m and we check whether this is correct
```

## Graphing a function

Now, we propose to plot the graph of a function. We have to use *matplotlib*:

```python
import matplotlib.pyplot as plt
import numpy as np

# We first define the x-interval and choose 100 linearly spaced values in the interval
x = np.linspace(-5,5,100)

# the function, which is y = x^2 here
y = x**2

# plot the function
plt.plot(x,y, 'r')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# We first define the x-interval and choose 100 linearly spaced values in the interval
x = np.linspace(-5,5,100)

# the function, which is y = x^2 here
y = x**2


# We can set the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'r')

# show the plot
plt.show()
```

We can also plot several functions on the same graph:

```python
import matplotlib.pyplot as plt
import numpy as np

# We first define the x-interval and choose 100 linearly spaced values in the interval
x = np.linspace(-3,3,100)

# three different functions
y1 = x**2
y2 = x**3
y3 = np.exp(x)


# We can set the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions
plt.plot(x,y1, 'r', label='y=x^2')
plt.plot(x,y2, 'b', label='y=x^3')
plt.plot(x,y3, 'g', label='y=e^x')
plt.legend(loc='upper left') # here is the legend using the label above. loc='position' allows to locate the legend on the graph

# show the plot
plt.show()
```
