# Simple interest

## Model

Imagine you have \$10,000 and you want to gain interests. You decide to go with a bank that offers *simple interest* with a \$150 of annual interest rate. This means that after the first year, we will get $10000+150 = \$10150$, after 2 years: $10150+150=\$10300$, and so on. 

*Arithmetic sequence* is the mathematical concept we use to model simple interest. We say *arithmetic* because we add a constant annual interest every year (which is like 1+1=2, 1+2=3, etc.. like arithmetics). 

Let's write the equations to model the balance in your bank account. We denote $u_0$ your initial balance ($u_0=10,000$ for our specific example), $u_1$ your balance after one year ($u_1=10,150$ for our specific example), $u_2$ your balance after two years ($u_1=10,300$ for our specific example), etc.., $u_k$ your balance after $k$ years ($u_k=10,000 + k\times 150$ for our specific example). In general, if your interest rate is $r$ (in math we also use $d$ for common difference) and your initial balance is $u_0$, then your balance after $k$ years will be:
$$
u_k = u_0 + k\ r\ .
$$
In math, we call this relation a *recursive formula* or *recurrence relation*. 
For our example, $u_0=10,000$ and $r=150$. 
In math, we call this relation the general term of the sequence. But we also use another relation that is more useful for computing purposes. The idea is the following: if I know my balance after $k$ years, I can calculate my balance in the subsequent year $k+1$ using the relation:
$$
u_{k+1} = u_{k} + r\ .
$$
I think you see where I'm going. If we use a *for loop* to iterate the calculation of the terms of the sequence. This is what we'll do next.


## Code

Let's print the list of the first 10 years:

```python
u=10000 # Initial condition
N=10
r=150
for n in range(N+1):
  print(n,u)
  u=u+r
  n=n+1
```

We could also use the while loop:

```python
u=10000 # Initial condition
N=10
r=150
n=0
while n<=10:
  print(n,u)
  u=u+r
  n=n+1
```

Or create a list:

```python
u=10000 # Initial condition
N=10
r=150
L=[]
for n in range(N+1):
  L.append(u)
  u=u+r
  n=n+1
print(L)
```

And graph it:

```python
from matplotlib import pyplot as plt
u=10000 # Initial condition
N=10
r=150
L=[]
for n in range(N+1):
  L.append(u)
  u=u+r
  n=n+1
plt.plot(L,'ro')


```

Say, we put our money in our bank account in 2022 and we want to graph the balance as a function of the year (i.e., 2023, 2024, etc..) instead of the number of years of saving. In this case, we should create a list $Lyear$ and graph the list of your balance $Lbalance$ versus $Lyear$.  

```python
from matplotlib import pyplot as plt
u=10000 
t=2022
N=10
r=150
Lbalance=[]
Lyear=[]
for n in range(N+1):
  Lbalance.append(u)
  Lyear.append(t)
  u=u+r
  t=t+1
  n=n+1
plt.plot(Lyear,Lbalance,'ro')
```

# Compound interest

## Model

Now you decide to save your \$10,000 in a bank that offers *compound interest* with $1\%$ annual interest rate. This means that after the first year, we will get $10000+0.01\times 10000 = 1.01\times 10000 = \$10100$, after 2 years:  $10100+0.01\times 10100 = 1.01\times 10100 = \$10201$, and so on. 

*Geometric sequence* is the mathematical concept we use to model simple interest. We say *geometric* because the ratio of two subsequent balance is constant, e.g., this ratio equals $1.01$ in the example above. 

Let's write the equations to model the balance in your bank account. We denote $u_0$ your initial balance ($u_0=10,000$ for our specific example), $u_1$ your balance after one year ($u_1=10,100$ for our specific example), $u_2$ your balance after two years ($u_1=10,201$ for our specific example), etc.., $u_k$ your balance after $k$ years. In general, if your interest rate is $r$ (in math we call the term $1+r$ a *common ratio*) and your initial balance is $u_0$, then your balance after $n$ years will be:
$$
u_n = u_0 \times (1+r)^n\ .
$$
For our example, $u_0=10,000$ and $r=0.01$. 
This relation gives the general term of the geometric sequence. To find the recursive formula, we simply use the fact that the common ratio between two subsequent terms is constant:
$$
\frac{u_{n+1}-u_n}{u_n} = r\ ,
$$
and hence, we obtain the following relation:
$$
u_{n+1} = u_{n} \times (1+r)
$$
As before, if we use a *for loop* to iterate the calculation of the terms of the sequence. This is what we'll do next.

## Code

```python
u=10000
N=10
r=0.01
for n in range(N+1):
  print(n,u)
  u=u*(1+r)
  n=n+1
```

We could also use the while loop:

```python
u=10000
N=10
r=0.01
n=0
while n<=10:
  print(n,u)
  u=u*(1+r)
  n=n+1
```

Or create a list:

```python
u=10000
N=10
r=0.01
L=[]
for n in range(N+1):
  L.append(u)
  u=u*(1+r)
  n=n+1
print(L)
```

And graph it (we now take 100 generations):

```python
from matplotlib import pyplot as plt
u=10000
N=100
r=0.01
L=[]
for n in range(N+1):
  L.append(u)
  u=u*(1+r)
  n=n+1
plt.plot(L,'ro')


```

# Comparison between the two models

## Graphing

```python
from matplotlib import pyplot as plt
u1=10000 # initial balance
u2=10000
t=2022
N=100 # number of years
r1=150 # simple interest rate
r2=0.01 # compound interest rate
L1=[] # List of simple interest account balances
L2=[] # List of compound interest account balances
Lt=[] # List of years
for n in range(N+1):
  L1.append(u1)
  L2.append(u2)
  Lt.append(t)
  u1=u1+r1
  u2=u2*(1+r2)
  t=t+1
  n=n+1
plt.plot(Lt,L1,'ro'), plt.plot(Lt,L2,'bo'), plt.legend(["Simple","Compound"])
```

## Finding the intercept

We want to know in which year the compound interest ccount balance will be greater than the simple interest one. 

To do this, we should use a *while loop*. Basically, we should iterate the loop until the value of the compound interest account balance is greater than the simple interest account balance. 

```python
u1=10000 # initial balance
u2=10000
t=2022
N=100 # number of years
r1=150 # simple interest rate
r2=0.01 # compound interest rate
while u2<=u1:
  u1=u1+r1
  u2=u2*(1+r2)
  t=t+1
  n=n+1
print("The compound interest account balance will be greater than the simple interest account balance in ",t)
```

```python
r0 = float(input("Taux d'intrt = "))
P0 = float(input("Solde initial (valeur de l'emprunt) = "))
N0 = int(input("Dure du remboursement (en mois) = "))
def MP(r,P,N):
    q=r/12
    W=(1+q)**N
    y=P*q*W/(W-1)
    return y

def bal(r,P,k,N):
    y=P
    q=r/12
    for i in range(1,k+1):
        y=(1+q)*y-MP(r,P,N)
        i=i+1
    return y

def Nmin(r,P,N):
    k=1
    while bal(r,P,k,N)>1*P/2:
        k=k+1
    return k

print("Votre versement mensuel est de ", MP(r0,P0,N0)),
print("Vous aurez rembours la moit de votre emprunt aprs ", Nmin(r0,P0,N0)," mois")
print("Vous aurez pay un total de ", MP(r0,P0,N0)*N0)
for k in range(1,N0+1):
    print(k,bal(r0,P0,k,N0))
```

```python
Max = float(input("Maximal monthly payment = "))
r0 = float(input("Interest rate = "))
V0 = float(input("Initial value of the car = "))
D0 = float(input("Deductible = "))
N0 = int(input("Duration of the loan (in months) = "))
P0 = V0-D0
def MP(r,P,N):
    q=r/12
    W=(1+q)**N
    y=P*q*W/(W-1)
    return y

def R(r,N):
    q=r/12
    W=(1+q)**N
    y=q*W/(W-1)
    return y

Dmin = V0 - Max/(R(r0,N0))

def rmax(r,P,N):
    x=r
    while MP(x,P,N)>Max:
        x=x-0.01/100
    return x

def Nmax(r,P,N):
    x=N
    while MP(r,P,x)>Max:
        x=x+1
    return x

if MP(r0,P0,N0)>=Max:
    print("Your monthly payment is ", MP(r0,P0,N0)), 
    print("Either the value of the principal is too high or the interest rate is too high or the duration of the loan is not long enough or the deductible is too low."),
    print("There are several possibilities:"),
    print("If your deductible was greater than ", Dmin,", your monthly payment would be less than ", Max),
    print("If the interest rate was less than ", rmax(r0,P0,N0),", your monthly payment would be less than ", Max),
    print("If the duration of the loan was greater than ", Nmax(r0,P0,N0),", your monthly payment would be less than ", Max)
else:
    print("Your monthly payment is ", MP(r0,P0,N0), "You can get this loan.")
```
