# Euclidean division

## Definition and code

Let $n$ and $b\neq 0$ be two integers. The Euclidean division is defined as follows:
$$n=bq+r$$
where $q\in\mathbb{Z}$ is the quotient and $0\geq r <|b|$ is the remainder. We say that $a$ is the divident and $b$ is the divisor. 

There are two operations in python to calculate the quotient $q$ and the remainder $r$:  
$$\text{Quotient: n//b = q}$$ 
$$\text{Remainder: n%q = r}$$

Let's look at the examples below. 

We first find the quotient for the following examples:

```python
print(3//2)
print(5//2)
print(25//2)
print(30//2)
print(343//13)
```

And now, let's find their reminder:

```python
print(3%2)
print(5%2)
print(25%2)
print(30%2)
print(343%13)
```

# Greatest common divisor - GCD

## Euclidean algorithm

Euclidean algorithm:

Let a and b two integers such that a ≥ b. To find their GCD, we'll use the Euclidean algorithm. 
*   We first calculate the remainder of a/b. Let's put $a_0=a,\ b_0=b$. We have: $$a_0=b_0 q_0+r_0$$, where $r_0$ is the remainder. 
*   If $r_0\neq0$, we put $a_1 = b_0$ and $b_1 = r_0$ and we repeat the process: $$a_1=b_1 q_1+r_1$$
*   We iterate the procedure for any $k\in\mathbb{N}$:
$$
a_{k+1}=b_{k+1}q_{k+1}+r_{k+1}\ ,
$$
where $a_{k+1}=b_{k}$ and $b_{k+1}=r_{r}$, until $r_{k+1}=0$. 
*   If for a rank $n+1$ we have $r_{n+1}=0$, then the GCD is nothing but the previous value of the remainder, i.e., $GCD = r_{n}$.





```python
a=eval(input("a= ")) 
b=eval(input("b= ")) 
r=a%b
while r!=0:
  d=r  
  a=b
  b=r
  r=a%b
print("GCD=",d)
if d==1:
  print("Those two integers are relatively primes (coprime)")
```

```python
n = int(input("Enter the number for calculating the prime factors :\n"))
for i in range(2,n + 1):
    if n % i == 0:
        count = 1
        for j in range(2,(i//2 + 1)):
            if(i % j == 0):
                count = 0
                break
        if(count == 1):
            print(i)
```

Now, we can find the prime decomposition:
$$n = p_1^{k_1}\times p_2^{k_2}\times \cdots\times p_{N_{f}}^{k_{N_{f}}} = \prod_{l=1}^{N_{f}} p_{i}^{k_i}$$

We have to find the number of occurence for each prime number $p_i$. 

First, let's write a code to find the list of the prime numbers with their respective exponents. The list will have the form: $[[p_2,k_1],[p2,k_2],\cdots,[p_{N_f},k_{N_f}]]$.

```python
n=343
L=[]
for i in range(2,n+1):
  if prime(i)==True:
    if n%i==0:
      s=0
      m=n
      while m%i==0:
        s=s+1
        m=m/i
      L.append([i,s])
print(L)
```

We can also create a function that return the list:

```python
def prime_decomp(n):
  L=[]
  for i in range(2,n+1):
    if prime(i)==True:
      if n%i==0:
        s=0
        m=n
        while m%i==0:
          s=s+1
          m=m/i
        L.append([i,s])
  return L
print(prime_decomp(343))
print(prime_decomp(1984))
```

# Collatz Conjecture

Consider an integer $n_0$. You decide to work with this number following simple rules: 
*   If $n$ is even, you divide the number by 2: $n_0\mapsto \frac{n_0}{2}$
*   If $n$ is odd, you multiply the number by 3 and you add 1 to this product: $n_0\mapsto 3n_0+1$

Then, you get a new number, say $n_1$, and you can repeat the same operation, and obtain a number $n_2$, and repeat the same operation over and over. Then you get a list of numbers: $n_0,n_1,n_2,\cdots, n_k,\cdots$. The question is: at the end of the day, will the sequence of numbers ever stop going? In other words, will the sequence oscillates between numbers indefinitely or will it converges to a number of a given sequence:
$$
n_0 \rightarrow n_1 \rightarrow n_2 \rightarrow n_3 \rightarrow \cdots \rightarrow n_{final}
$$

Let's start with an example. Take $n_0=7$. As $7$ is odd, we use the second rule: $n_1=3\times 7+1 = 22$. As $22$ is even, we use the first rule: $n_2 = 22/2=11$. After 16 iterations, the number will be equal to 1:
$$
7 → 22 → 11 → 34 → 17 → 52 → 26 → 13→ 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1   
$$
But once a number in the sequence equals 1, the sequence is stuck in a loop:
$$
1 → 4 → 2 → 1 
$$
So, we can consider that the sequence "converges" to 1. 

Collatz conjecture states that for any initial number $n_0$, the sequence will always converges to 1.

Formally, the problem can be written as follows: Let $n_0\in \mathbb{N}, n_0\geq 1$ and let $(u_n)$ be a sequence defined by the recursive formula:
$$
u_{n+1} = \begin{cases} 2u_n,\ \text{if } u_n\ \text{is even} \\ 3u_n+1,\ \text{if } u_n\ \text{is odd} \end{cases}\ ,\ \text{with  }u_{0}=n_0
$$
Then, $\forall n_0\in\mathbb{N}^{\ast}$, there exists a finite integer $N$ such that $u_N=1$. 

In this section, we propose to verify that it works numerically. 

## Print the list of number in the sequence 

Let's print the list of the numbers in the sequence for $n_0=7$:

```python
n=7
L=[n]
while n!=1:
  if n%2 != 0:
    n=3*n+1
  else:
    n=n/2
  L.append(n)
print(L)
```

# Goldbach conjecture

## Conjecture

The Goldbach conjecture states that for all number $n\geq 2$, there exists two prime numbers $p_1$ and $p_2$ so that:
$$
2n=p_1 + p_2\ .
$$
It looks obvious after you try with a couple of numbers, and yet, it has never been proven. The conjecture was formulated the first time by a German mathematician Christian Goldbach in the 18th century. It is still an open problem in mathematics.

## Algorithm to determine if a number is prime or not

I propose a code to check whether a number is prime: 

```python
def prime(num):
    if num> 1:  
        for n in range(2,num):  
            if (num % n) == 0:  
                return False
        return True
    else:
        return False
print(prime(13))
print(prime(64))
print(prime(5))
```

## Algorithm to check the Goldbach conjecture

```python
n=25
s=0
for k in range(n+1):
  n1=2*n-(2*k+1)
  n2=2*n+(2*k+1)
  if prime(n1)==False or prime(n2)==False:
    k=k+1
  else:
    s=s+1
    print("true ,",2*n,"=",n1,"+",n2)
print("There are ",s," possibilities")
    
```
